```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset (headers are often missing in this CSV, so we add them)
columns = ['ID', 'Entity', 'Sentiment', 'Tweet_Content']
df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Documents\\MUMBAI UNIVERSITY\\SEM 2\\PRODIGY_DS_04\\twitter_training.csv", names=columns)

# 2. Clean the data (Drop rows where the tweet is empty)
df.dropna(subset=['Tweet_Content'], inplace=True)

# 3. Check the distribution of sentiments
sentiment_counts = df['Sentiment'].value_counts()
print(sentiment_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution of Twitter Data')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.savefig('sentiment_distribution.png')
plt.show()
```

```python
# Filter for a specific entity, e.g., 'Microsoft'
entity_df = df[df['Entity'] == 'Microsoft']
entity_sentiment = entity_df['Sentiment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(entity_sentiment, labels=entity_sentiment.index, autopct='%1.1f%%', colors=['green', 'red', 'gray', 'blue'])
plt.title('Sentiment Analysis for Microsoft')
plt.savefig('microsoft_sentiment.png')
plt.show()
```
